```python
from PIL import Image
import os
import matplotlib.pyplot as plt
```

```python
def get_file_name(directory):
    file_names = []
    for file in os.listdir(directory):
        name_without_ext = os.path.splitext(file)[0]
        file_names.append(name_without_ext) 
    return file_names
```

```python
# Get the paths of the images in different directories
def get_images_paths(file_names, directories):
    images_paths = []
    for file_name in file_names:    
        image_paths = []        
        for directory, extension in directories.items():   
            file = f"{file_name}.{extension}"
            image_paths.append(os.path.join(directory, file))
        images_paths.append(image_paths)
    return images_paths
```

```python
def show_depth_images(images_paths):
    for image_paths in images_paths:
        plt.figure(figsize=(15, 10))     
        for i, image_path in enumerate(image_paths):
            image = Image.open(image_path)
            if i == 0:
                image = image.rotate(270, expand=True)

            n_cols = (len(image_paths) + 1) // 2  
            plt.subplot(2, n_cols, i + 1)
            plt.imshow(image)
            plt.title(os.path.basename(os.path.dirname(image_path)))
            plt.axis('off')
            
        plt.show()
```

### Show Depth Map Prediction of 0.5m_with_scale

```python
directories = {
    "Depth Map Comparison/0.5m_with_scale/": "jpg",
    "Depth Map Comparison/0.5m_with_scale_pred//depth_anything_vits/": "png",
    "Depth Map Comparison/0.5m_with_scale_pred/depth_anything_vitb/": "png",
    "Depth Map Comparison/0.5m_with_scale_pred/depth_anything_vitl/": "png",
    "Depth Map Comparison/0.5m_with_scale_pred/depth_anything_metric_vits/": "png",
    "Depth Map Comparison/0.5m_with_scale_pred/depth_anything_metric_vitb/": "png",
    "Depth Map Comparison/0.5m_with_scale_pred/depth_anything_metric_vitl/": "png",
}

show_depth_images(get_images_paths(get_file_name('Depth Map Comparison/0.5m_with_scale'), directories))
```
