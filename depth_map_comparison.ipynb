```python
from PIL import Image
import os
import matplotlib.pyplot as plt

DIRECTORIES = {
    "0.5m_with_scale/": "jpg",
    "Depth-Anything-V2/depth_anything_vits/": "png",
    "Depth-Anything-V2/metric_depth/depth_anything_metric_vits/": "png",
}
```

```python
def get_file_name(directory):
    file_names = []
    for file in os.listdir(directory):
        name_without_ext = os.path.splitext(file)[0]
        file_names.append(name_without_ext) 
    return file_names
get_file_name('0.5m_with_scale')
```

```python
# Get the paths of the images in different directories
def get_images_paths(file_names):
    images_paths = []
    for file_name in file_names:    
        image_paths = []        
        for directory, extension in DIRECTORIES.items():   
            file = f"{file_name}.{extension}"
            image_paths.append(os.path.join(directory, file))
        images_paths.append(image_paths)
    return images_paths

get_images_paths(get_file_name('0.5m_with_scale'))
```

```python
from PIL import Image
import os

DIRECTORIES = {
    "0.5m_with_scale/": "jpg",
    "Depth-Anything-V2/vis_depth/": "png",
    "Depth-Anything-V2/metric_depth/vis_depth/": "png",
}
```

```python
def get_file_name(directory):
    file_names = []
    for file in os.listdir(directory):
        name_without_ext = os.path.splitext(file)[0]
        file_names.append(name_without_ext) 
    return file_names
get_file_name('0.5m_with_scale')
```

```python
# Get the paths of the images in different directories
def get_images_paths(file_names):
    images_paths = []
    for file_name in file_names:    
        image_paths = []        
        for directory, extension in DIRECTORIES.items():   
            file = f"{file_name}.{extension}"
            image_paths.append(os.path.join(directory, file))
        images_paths.append(image_paths)
    return images_paths

get_images_paths(get_file_name('0.5m_with_scale'))
```

```python
def show_depth_images(images_paths):
    for image_paths in images_paths[0:5]:
        plt.figure(figsize=(15, 5))     
        for i, image_path in enumerate(image_paths):
            image = Image.open(image_path)
            if i == 0:
                image = image.rotate(270, expand=True)
            plt.subplot(1, len(images_paths[0]), i + 1)
            plt.imshow(image)
            plt.title(os.path.basename(os.path.dirname(image_path)))
            plt.axis('off')
        plt.show()
        
show_depth_images(get_images_paths(get_file_name('0.5m_with_scale')))
```
